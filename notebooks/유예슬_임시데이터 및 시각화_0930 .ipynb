```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.metrics import jaccard_score
bert_model = SentenceTransformer('jhgan/ko-sroberta-multitask')

def evaluate_similarity(label_text, review_text, alpha=0.9): # cosine 가중치 조절
    label_embedding = bert_model.encode(label_text)
    review_embedding = bert_model.encode(review_text)

    # 코사인 유사도
    cosine_sim = 1 - pairwise_distances([label_embedding], [review_embedding], metric='cosine')[0][0]

    # 자카드 유사도
    label_tokens = set(label_text.split())
    review_tokens = set(review_text.split())
    jaccard_sim = len(label_tokens.intersection(review_tokens)) / len(label_tokens.union(review_tokens))

    # 가중 평균 내보기
    weighted_sim = alpha*cosine_sim + (1 - alpha)*jaccard_sim

    return weighted_sim
```

```python
temporary_pos['date'] = pd.to_datetime(temporary_pos['date'])
temporary_pos = temporary_pos.set_index('date')

pos_7D_score = temporary_pos['score'].resample('7D',closed='left').sum().reset_index()
pos_7D_score
```

```python
temporary_neg['date'] = pd.to_datetime(temporary_neg['date'])
temporary_neg = temporary_neg.set_index('date')

neg_7D_score = temporary_neg['score'].resample('7D',closed='left').sum().reset_index()
neg_7D_score
```

```python
merged_temporary_df = pd.merge(pos_7D_score, neg_7D_score, on='date')
merged_temporary_df
```

```python
merged_temporary_df.columns = ['date','pos_7D_score','neg_7D_score' ]
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))  # 그래프 크기 설정

# 꺾은선 그래프 그리기
sns.lineplot(data=merged_temporary_df, x='date', y='pos_7D_score', label='Positive', marker='o', linestyle='-', color='b')
sns.lineplot(data=merged_temporary_df, x='date', y='neg_7D_score', label='Negative', marker='s', linestyle='--', color='r')

# 그래프 면적을 색으로 채우기 (양의 부분, 녹색)
plt.fill_between(merged_temporary_df['date'], merged_temporary_df['pos_7D_score'], color='#a8dadc', alpha=0.3, label='Positive Area')

# 그래프 면적을 색으로 채우기 (음의 부분, 붉은색)
plt.fill_between(merged_temporary_df['date'],merged_temporary_df['neg_7D_score'], color='#f1a0a0', alpha=0.3, label='Negative Area')

# 그래프 제목과 레이블 설정
plt.title('긍부정 지수 (2개월)', fontsize=14)
plt.xlabel('일자', fontsize=12)
plt.ylabel('긍부정 지수', fontsize=12)

# 범례 추가
plt.legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python

# 스케일링
def scale_minmax(merged_temporary_df):
    min_value = min(merged_temporary_df)
    size_value = max(merged_temporary_df) - min_value
    return [ (x-min_value)/size_value for x in merged_temporary_df]

pos_7D_score['score'] = scale_minmax(pos_7D_score['score'])
neg_7D_score['score'] = scale_minmax(neg_7D_score['score'])
pos_7D_score
```

### 긍부정 지수변화 (전체기간) ***

```python
import numpy as np
from scipy.interpolate import make_interp_spline

# 기존 데이터 포인트
x = merged_df['DATE']
y_pos = merged_df['pos_7D_score']
y_neg = merged_df['neg_7D_score']

# 날짜 데이터를 숫자로 변환
x_numeric = pd.to_numeric(x)

# 더 작은 간격으로 나눈 새로운 데이터 포인트
x_new = np.linspace(x_numeric.min(), x_numeric.max(), 300)  # 300개의 데이터 포인트로 나눔

# 스플라인 곡선 생성
spline_pos = make_interp_spline(x_numeric, y_pos, k=3)  # k는 스플라인의 차수를 나타냄
spline_neg = make_interp_spline(x_numeric, y_neg, k=3)

# 부드러운 스플라인 곡선 값 계산
y_pos_smooth = spline_pos(x_new)
y_neg_smooth = spline_neg(x_new)

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(x_new, y_pos_smooth, label='긍정 지수', color='b')
plt.plot(x_new, y_neg_smooth, label='부정 지수', color='r')

# 그래프 면적을 색으로 채우기 (양의 부분, 녹색)
plt.fill_between(x_new, y_pos_smooth, color='#a8dadc', alpha=0.3, label='Positive Area')

# 그래프 면적을 색으로 채우기 (음의 부분, 붉은색)
plt.fill_between(x_new, y_neg_smooth, color='#f1a0a0', alpha=0.3, label='Negative Area')

# 그래프 제목과 레이블 설정
plt.title('긍부정 지수 변화 (전체기간)', fontsize=14)
plt.xlabel('일자', fontsize=12)
plt.ylabel('리뷰 긍부정 지수', fontsize=12)

# 범례 추가
plt.legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import ttest_ind

# 음수 평균 점수와 양수 평균 점수를 각각의 그룹으로 나누기
negative_scores = final_result_id_df2[final_result_id_df2['MEAN_NEW_SCORE'] < 0]['id_count']
positive_scores = final_result_id_df2[final_result_id_df2['MEAN_NEW_SCORE'] >= 0]['id_count']

# t-검정 수행
t_stat, p_value = ttest_ind(negative_scores, positive_scores)

# 결과 출력
print(f't-검정 통계량: {t_stat}')
print(f'p-value: {p_value}')

# p-value를 기준으로 유의미성 판단
alpha = 0.05
if p_value < alpha:
    print('음수 평균 점수와 양수 평균 점수 간의 차이는 유의미합니다.')
else:
    print('음수 평균 점수와 양수 평균 점수 간의 차이는 유의미하지 않습니다.')
```

```python
tmp_df1 = df1.groupby('ID').size().reset_index(name='id_count').sort_values(by='id_count', ascending=False)

# 'id' 별로 리뷰를 남긴 가장 이른 날짜와 가장 늦은 날짜 찾기
df1['DATE'] = pd.to_datetime(df1['DATE'])
review_periods = df1.groupby('ID')['DATE'].agg(['min', 'max'])

# 각 고객의 활성 리뷰 기간 계산 (가장 늦은 날짜 - 가장 이른 날짜)
review_periods['active_days'] = (review_periods['max'] - review_periods['min']).dt.days
review_periods

tmp_df1 = pd.merge(tmp_df1, review_periods, on='ID')

#긍부정 리뷰횟수

positive_num = df1.groupby('ID')['SENTIMENT'].sum()
tmp_df1 = pd.merge(tmp_df1, positive_num, on='ID').rename(columns={'SENTIMENT': 'pos_num'})

tmp_df1['neg_num'] = tmp_df1['id_count'] - tmp_df1['pos_num']
tmp_df1

# 온도 score 누적

df1['NEW_SCORE'] = df1.apply(lambda row: row['SCORE'] if row['SENTIMENT'] == 1 else - row['SCORE'], axis=1)
df1 # new score 추가함

score_sum = df1.groupby('ID')['NEW_SCORE'].sum()
score_sum

tmp_df1 = pd.merge(tmp_df1, score_sum, on='ID').rename(columns={'NEW_SCORE': 'score_sum'})
tmp_df1

#온도 평균, 중앙값

score_mean = df1.groupby('ID')['NEW_SCORE'].mean()
score_mean

tmp_df1 = pd.merge(tmp_df1, score_mean, on='ID').rename(columns={'NEW_SCORE': 'score_mean'})
tmp_df1

score_median = df1.groupby('ID')['NEW_SCORE'].median()
score_median

tmp_df1 = pd.merge(tmp_df1, score_median, on='ID').rename(columns={'NEW_SCORE': 'score_median'})
tmp_df1

# 온도 min max 값

score_min = df1.groupby('ID')['NEW_SCORE'].min()
score_min

tmp_df1 = pd.merge(tmp_df1, score_min, on='ID').rename(columns={'NEW_SCORE': 'score_min'})
tmp_df1

# 온도 max값
score_max = df1.groupby('ID')['NEW_SCORE'].max()
score_max

tmp_df1 = pd.merge(tmp_df1, score_max, on='ID').rename(columns={'NEW_SCORE': 'score_max'})
tmp_df1

# 온도 표준편차값
score_std = df1.groupby('ID')['NEW_SCORE'].std()
score_std

tmp_df1 = pd.merge(tmp_df1, score_std, on='ID').rename(columns={'NEW_SCORE': 'score_std'})
tmp_df1
```
