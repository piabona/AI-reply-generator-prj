```python
from kiwipiepy import Kiwi
from sklearn.feature_extraction.text import TfidfVectorizer

# 텍스트 데이터
texts = train['고객리뷰']

# Kiwi 형태소 분석기 초기화
kiwi = Kiwi()

```

```python
train['']
```

```python
# 형태소 분석 및 명사 추출
keyword_texts = []
for text in texts:
    kiwi_result = kiwi.analyze(text)
    keyword = [word[0] for word in kiwi_result[0][0] if word[1] == ["NNG",'VA','XR']]  # 명사만 추출
    keyword_texts.append(" ".join(keyword))

```

```python
keyword_texts
```

```python
# TF-IDF 벡터화
tfidf_vectorizer = TfidfVectorizer(max_features=10)  # 가장 빈도수 높은 명사 10개만 선택
tfidf_matrix = tfidf_vectorizer.fit_transform(keyword_texts)

# TF-IDF 값 출력
feature_names = tfidf_vectorizer.get_feature_names_out()
for i, text in enumerate(texts):
    tfidf_values = tfidf_matrix[i].toarray()[0]
    sorted_indices = tfidf_values.argsort()[::-1]
    print(f"Text {i + 1}:")
    for idx in sorted_indices:
        print(f"{feature_names[idx]}: {tfidf_values[idx]:.4f}")
    print()
```
