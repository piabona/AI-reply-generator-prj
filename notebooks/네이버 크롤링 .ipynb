```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pandas as pd

url = "https://pcmap.place.naver.com/restaurant/1050919643/review/visitor?entry=pll&from=map&fromPanelNum=1&ts=1692159860047"
browser.get(url)

browser.implicitly_wait(2)
time.sleep(3)

# ë” ì´ìƒ ìŠ¤í¬ë¡¤ì´ ë˜ì§€ ì•Šì„ ë•Œê¹Œì§€ ìŠ¤í¬ë¡¤ ë°˜ë³µ
last_height = browser.execute_script("return document.body.scrollHeight")
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)  # ìŠ¤í¬ë¡¤ í›„ ì½˜í…ì¸  ë¡œë”© ëŒ€ê¸°

    # "í¼ì³ë³´ê¸°" ë²„íŠ¼ì„ ì°¾ì•„ì„œ í´ë¦­
    expand_buttons = browser.find_elements(By.CSS_SELECTOR, '.P1zUJ.ZGKcF')
    for button in expand_buttons:
        action = ActionChains(browser)
        action.move_to_element(button).click(button).perform()
        time.sleep(1)  # í´ë¦­ í›„ ì½˜í…ì¸  ë¡œë”© ëŒ€ê¸°

    # ìŠ¤í¬ë¡¤ì´ ë” ì´ìƒ ë˜ì§€ ì•ŠëŠ”ì§€ í™•ì¸
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

reviews = []

review_elements = browser.find_elements(By.CSS_SELECTOR, 'li.YeINN')
for review_element in review_elements:
    html = review_element.get_attribute('outerHTML')
    soup = BeautifulSoup(html, 'html.parser')

    username = soup.select_one('.VYGLG').text
    review_text = soup.select_one('.ZZ4OK .zPfVt').text
    visit_date = soup.select_one('.tzZTd time').text
    visit_count_text = soup.select('.tzZTd')[1].text
    visit_count = visit_count_text.replace("ë²ˆì§¸ ë°©ë¬¸", "").strip()

    positive_points_elements = soup.select('.gyAGI .P1zUJ')
    positive_points = [elem.text for elem in positive_points_elements]

    review_data = {
        'ì•„ì´ë””': username,
        'ë¦¬ë·°': review_text,
        'ë°©ë¬¸ì¼': visit_date,
        'ë°©ë¬¸íšŸìˆ˜': visit_count,
        'ì´ëŸ°ì ì´ ì¢‹ì•˜ì–´ìš”': ', '.join(positive_points)
    }
    reviews.append(review_data)

browser.quit()

df = pd.DataFrame(reviews)

filename = "naver_reviews11.xlsx"
df.to_excel(filename, index=False, engine='openpyxl')

print(f"{filename}ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
